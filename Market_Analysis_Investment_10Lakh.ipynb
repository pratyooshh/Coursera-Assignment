## 1. 📉 Current Market Conditions — March 2026

The Indian equity market is going through a **bearish and volatile phase** in March 2026. Here's a snapshot of the key drivers:

```python
# Current Market Snapshot — March 2026
market_data = {
    'Index': ['Nifty 50', 'Sensex', 'Nifty Bank', 'Nifty IT', 'Nifty Midcap 100', 'Nifty Smallcap 100'],
    'Current Level': [23151, 76704, 53757, 36800, 51200, 16800],
    '52W High': [26277, 85978, 54467, 44900, 60000, 21000],
    '52W Low': [21964, 72685, 48929, 33800, 45500, 14600],
    'Weekly Change (%)': [-5.31, -3.20, -6.97, -4.10, -5.50, -6.80],
    'YTD Change (%)': [-8.20, -7.80, -9.50, -11.20, -10.40, -14.20]
}

df_market = pd.DataFrame(market_data)
df_market.set_index('Index', inplace=True)
print('=== Indian Market Snapshot — March 2026 ===')
print(df_market.to_string())
```

```python
# Visualize YTD performance
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# YTD Change
colors = ['red' if x < 0 else 'green' for x in df_market['YTD Change (%)']]
axes[0].barh(df_market.index, df_market['YTD Change (%)'], color=colors, edgecolor='black', alpha=0.8)
axes[0].axvline(0, color='black', linewidth=1.5)
axes[0].set_title('YTD Performance — Indian Indices (2026)', fontweight='bold', fontsize=13)
axes[0].set_xlabel('YTD Change (%)')
for i, (idx, val) in enumerate(zip(df_market.index, df_market['YTD Change (%)'])):
    axes[0].text(val - 0.3, i, f'{val}%', va='center', ha='right', fontsize=10, color='white', fontweight='bold')

# 52W Range
for i, idx in enumerate(df_market.index):
    low = df_market.loc[idx, '52W Low']
    high = df_market.loc[idx, '52W High']
    current = df_market.loc[idx, 'Current Level']
    pct_from_high = ((current - high) / high) * 100
    axes[1].barh(i, pct_from_high, color='tomato', alpha=0.7, edgecolor='black')
    axes[1].text(pct_from_high - 0.5, i, f'{pct_from_high:.1f}%', va='center', ha='right', fontsize=9, color='white', fontweight='bold')

axes[1].set_yticks(range(len(df_market.index)))
axes[1].set_yticklabels(df_market.index)
axes[1].set_title('Distance from 52-Week High (%)', fontweight='bold', fontsize=13)
axes[1].set_xlabel('% Below 52W High')
axes[1].axvline(0, color='black', linewidth=1.5)

plt.tight_layout()
plt.savefig('market_snapshot.png', dpi=150, bbox_inches='tight')
plt.show()
print('Chart saved as market_snapshot.png')
```

## 4. 💼 Recommended Portfolio Allocation — ₹10 Lakh

Based on current market conditions (bearish correction + long-term India growth story), here are **three portfolio strategies** for different risk profiles:

```python
TOTAL = 1_000_000  # 10 Lakh

# ---- Portfolio Definitions ----
portfolios = {
    'Conservative (Low Risk)': {
        'PPF / Govt Bonds': 0.25,
        'Fixed Deposit': 0.20,
        'Debt Mutual Fund': 0.15,
        'Large Cap Index Fund': 0.20,
        'Sovereign Gold Bonds': 0.15,
        'Emergency Liquid Fund': 0.05
    },
    'Balanced (Moderate Risk)': {
        'Large Cap Index Fund': 0.30,
        'Flexi Cap Mutual Fund': 0.20,
        'Mid Cap Mutual Fund': 0.15,
        'Sovereign Gold Bonds': 0.15,
        'PPF': 0.10,
        'InvITs / REITs': 0.10
    },
    'Aggressive (High Growth)': {
        'Large Cap Index Fund': 0.25,
        'Mid Cap Mutual Fund': 0.20,
        'Small Cap Mutual Fund': 0.15,
        'Direct Equity (Stocks)': 0.20,
        'Gold ETF': 0.10,
        'InvITs / REITs': 0.10
    }
}

# Expected returns for each asset
returns = {
    'PPF / Govt Bonds': 7.1, 'Fixed Deposit': 7.5, 'Debt Mutual Fund': 7.8,
    'Large Cap Index Fund': 12, 'Sovereign Gold Bonds': 8.5, 'Emergency Liquid Fund': 6.5,
    'Flexi Cap Mutual Fund': 14, 'Mid Cap Mutual Fund': 16, 'PPF': 7.1,
    'InvITs / REITs': 10, 'Small Cap Mutual Fund': 18, 'Direct Equity (Stocks)': 15,
    'Gold ETF': 8
}

print(f'{'='*60}')
print(f'  PORTFOLIO ALLOCATION — ₹{TOTAL:,.0f} (10 Lakh)')
print(f'{'='*60}')

for pf_name, alloc in portfolios.items():
    print(f'\n--- {pf_name} ---')
    expected_annual_return = 0
    for asset, pct in alloc.items():
        amount = TOTAL * pct
        exp_ret = returns.get(asset, 8)
        expected_annual_return += pct * exp_ret
        print(f'  {asset:<30s}  {pct*100:.0f}%   ₹{amount:>9,.0f}   (~{exp_ret}% p.a.)')
    print(f'  {"":<30s}  {100:.0f}%   ₹{TOTAL:>9,.0f}')
    print(f'  Expected Blended Annual Return: {expected_annual_return:.2f}%')
    print(f'  Expected Value in 5 Years:      ₹{TOTAL * ((1 + expected_annual_return/100)**5):,.0f}')
```

```python
# Pie charts for all 3 portfolios
fig, axes = plt.subplots(1, 3, figsize=(20, 7))

color_palettes = [
    ['#3498db', '#2980b9', '#1abc9c', '#16a085', '#f39c12', '#e67e22'],
    ['#2ecc71', '#27ae60', '#e74c3c', '#c0392b', '#9b59b6', '#8e44ad'],
    ['#e74c3c', '#c0392b', '#f39c12', '#d35400', '#1abc9c', '#3498db']
]

for ax, (pf_name, alloc), colors in zip(axes, portfolios.items(), color_palettes):
    labels = list(alloc.keys())
    sizes = [v * 100 for v in alloc.values()]
    wedges, texts, autotexts = ax.pie(
        sizes, labels=None, autopct='%1.0f%%',
        colors=colors, startangle=90,
        pctdistance=0.75, wedgeprops=dict(edgecolor='white', linewidth=2)
    )
    for at in autotexts:
        at.set_fontsize(10)
        at.set_fontweight('bold')
        at.set_color('white')
    ax.set_title(pf_name, fontweight='bold', fontsize=11)
    ax.legend(labels, loc='lower center', bbox_to_anchor=(0.5, -0.28), fontsize=7.5, ncol=2)

plt.suptitle('Portfolio Allocation — ₹10 Lakh Investment (March 2026)', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('portfolio_allocation.png', dpi=150, bbox_inches='tight')
plt.show()
print('Chart saved as portfolio_allocation.png')
```

## 5. 📈 Wealth Growth Projection — 10 Year Horizon

```python
# Wealth Growth Projections
years = np.arange(0, 11)

portfolio_returns = {
    'Conservative (8.2% p.a.)': 0.082,
    'Balanced (11.3% p.a.)': 0.113,
    'Aggressive (13.7% p.a.)': 0.137,
    'FD / Savings (7.5% p.a.)': 0.075  # baseline comparison
}

colors_proj = ['#3498db', '#2ecc71', '#e74c3c', '#95a5a6']
line_styles = ['-', '-', '-', '--']

fig, ax = plt.subplots(figsize=(12, 7))

for (name, rate), color, ls in zip(portfolio_returns.items(), colors_proj, line_styles):
    values = TOTAL * (1 + rate) ** years
    ax.plot(years, values / 1e5, label=name, color=color, linewidth=2.5, linestyle=ls, marker='o', markersize=5)
    ax.annotate(f'₹{values[-1]/1e5:.1f}L', (years[-1], values[-1]/1e5),
                textcoords='offset points', xytext=(5, 0), fontsize=9, color=color, fontweight='bold')

ax.axhline(TOTAL / 1e5, color='gray', linestyle=':', alpha=0.6, label='Initial Investment (₹10L)')
ax.fill_between(years, TOTAL / 1e5, portfolio_returns['Aggressive (13.7% p.a.)'] and
                TOTAL * (1 + 0.137) ** years / 1e5, alpha=0.05, color='green')

ax.set_xlabel('Years', fontsize=12)
ax.set_ylabel('Portfolio Value (₹ Lakhs)', fontsize=12)
ax.set_title('Wealth Growth Projection — ₹10 Lakh Investment Over 10 Years', fontweight='bold', fontsize=13)
ax.legend(fontsize=10)
ax.set_xticks(years)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'₹{x:.0f}L'))

plt.tight_layout()
plt.savefig('wealth_projection.png', dpi=150, bbox_inches='tight')
plt.show()
print('Chart saved as wealth_projection.png')
```

```python
# Detailed 10-Year Projection Table
print(f'\n{"="*70}')
print(f'  WEALTH GROWTH PROJECTION TABLE — ₹10 Lakh Investment')
print(f'{"="*70}')
print(f'{"Year":<8} {"Conservative":>16} {"Balanced":>14} {"Aggressive":>14} {"FD/Savings":>14}')
print(f'{"":<8} {"(8.2% p.a.)":>16} {"(11.3% p.a.)":>14} {"(13.7% p.a.)":>14} {"(7.5% p.a.)":>14}')
print('-' * 70)

for y in range(0, 11):
    vals = {name: TOTAL * (1 + r) ** y for name, r in portfolio_returns.items()}
    print(f'{y:<8d}', end='')
    for v in vals.values():
        print(f'  ₹{v/100000:>8.2f}L', end='')
    print()
```
